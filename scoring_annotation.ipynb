# formatting of the datafile

open sirius file compound_identifications.tsv and add the suffix _sirius to each column of the file

```python
df_sirius = pd.read_csv(input_sirius_path, sep='\t') 
suffix = "_sirius"

#adding _sirius suffix
for col in df_sirius.columns:
    if col != "id":
        new_col = col + suffix
        df_sirius.rename(columns={col: new_col}, inplace=True)

#for each element in column id, keep only the characters after the last '_'
df_sirius['id'] = df_sirius['id'].str.split('_').str[-1]
df_sirius["id"] = pd.to_numeric(df_sirius["id"], downcast='integer')



df_sirius.head(10)
```

Clean sirius file 

```python
# List of columns to keep
columns_to_keep = ['id', 'ConfidenceScore_sirius', 'CSI:FingerIDScore_sirius', 'SiriusScore_sirius', 'molecularFormula_sirius', 'adduct_sirius', 'InChIkey2D_sirius', 'InChI_sirius', 'name_sirius', 'smiles_sirius', 'xlogp_sirius', 'ionMass_sirius']

# Drop unwanted columns
columns_to_drop = [col for col in df_sirius.columns if col not in columns_to_keep]
df_sirius = df_sirius.drop(columns_to_drop, axis=1)

# Sirius file is cleaned !
```

Cleaned Met_annot_enhancer file

```python
df_metannot= pd.read_csv(input_metannot_path, sep='\t')
suffix = "_metAnnot"

for col in df_metannot.columns:
    if col != "feature_id":
        new_col = col + suffix
        df_metannot.rename(columns={col: new_col}, inplace=True)

df_metannot.head(10)
```

GNPS file 

```python
#open DB_result corresponding to your gnps job (csv)
df_gnps = pd.read_csv(input_gnps_path, sep='\t') 
suffix = "_GNPS"

for col in df_gnps.columns:
    if col != "#Scan#":
        new_col = col + suffix
        df_gnps.rename(columns={col: new_col}, inplace=True)
df_gnps.head()
```

Merging 

```python
# Merge the dataframes two by two based on the "id" or "feature_id" column
merged_df = pd.merge(df_sirius, df_metannot, left_on='id', right_on='feature_id', how='outer')

merged_df = pd.merge(merged_df, df_gnps, left_on='id', right_on='#Scan#', how='outer')

merged_df.head(10)
```
